```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Admin\Downloads\archive.zip")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df['runtime'].fillna(df['runtime'].median(), inplace=True)
df['vote_average'].fillna(df['vote_average'].mean(), inplace=True)
df.dropna(subset=['title'], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.figure()
plt.hist(df['vote_average'], bins=10, color='mediumslateblue', edgecolor='black')
plt.title("Distribution of Movie Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Movies")
plt.show()
```

```python
plt.figure()
plt.hist(df['runtime'], bins=15, color='seagreen', edgecolor='black')
plt.title("Movie Runtime Distribution")
plt.xlabel("Runtime (minutes)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure()
plt.scatter(df['popularity'], df['vote_average'], 
            color='tomato', alpha=0.6)
plt.title("Popularity vs Movie Rating")
plt.xlabel("Popularity")
plt.ylabel("Rating")
plt.show()
```
